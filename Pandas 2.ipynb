```python
users = pd.DataFrame({"userid":[1, 2, 3], "name":["sharadh", "shahid", "khusalli"]})
users
```

```python
msgs = pd.DataFrame({"userid":[1, 1, 2, 4], "msg":['hmm', "acha", "theek hai", "nice"]})
msgs
```

```python
# in numpy what's default value for axis -1 the largest axis available.
# In pandas, what's defautl value  for axis
import pandas as pd
t=pd.concat([users, msgs])
```

```python
import pandas as pd

# Original DataFrame from dictionary
df_dict = {'name': ['a', 'b', 'c', 'd'], 'age': [12, 15, 18, 19]}
df = pd.DataFrame(df_dict)

# List to be converted into a DataFrame
new_data = [['d', 20]]
new_df = pd.DataFrame(new_data, columns=['name', 'age'])

# Concatenate DataFrames
result = pd.concat([df, new_df], ignore_index=True)


result = result.drop(3)

print(result)
```
