```python
# np.where

# Product prices
prices = np.array([45, 55, 60, 75, 40, 90])

# apply 10% discount if price is above $50, otherwise do nothing
np.where(prices>60,prices*0.5,prices)
# np.where (if condition, what to do if condition satisfied, otherwise)
```
