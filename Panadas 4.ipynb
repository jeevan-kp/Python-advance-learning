```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('movies.csv',index_col=0)
directors = pd.read_csv('directors.csv', index_col=0)
movies
```

```python
data=movies.merge(directors,left_on='director_id',right_on='id',how='left')
data
```

```python
data.drop(['director_id','id_y'], axis=1, inplace=True)
data.head()
```

```python
data.groupby('director_name')['title'].count().sort_values(ascending=False)
```

```python
data_agg=data.groupby('director_name').agg({'year':['min','max'],"title":"count"})
data_agg
```

```python
data_agg.columns
```

```python
data_agg.reset_index()
```

```python
data_agg.columns
```

```python
data_agg.columns=["_".join(x) for x in data_agg.columns]
data_agg.reset_index()
```

```python
# Using the new features, can we find the most productive director?
# 1. first find out years active, then find out total no of movies he has made
# then years_active / total movies
data_agg['years_active']=data_agg['year_max']-data_agg['year_min']
data_agg.head()
```

```python
(data_agg['title_count']/data_agg['years_active']).sort_values(ascending=False)
```

```python
data = pd.read_csv('Pfizer_1.csv')
data.head(10)
```

# **Melt**

```python
data_melt=pd.melt(data,id_vars=['Date','Drug_Name','Parameter'],var_name='time',value_name='reading')
```

```python
data_melt.head()
```

```python
 data_melt = pd.melt(data, id_vars=['Date', 'Parameter', 'Drug_Name'], var_name = 'Drug_name', value_name='reading')
 data_melt.head()
```

# **Pivot**

```python
ano=data_melt.pivot(index=['Date', 'Parameter', 'Drug_Name'],columns='time',values='reading')
ano
```

```python
ano.reset_index()
```

```python
ano.columns.name=None
```

```python
ano.head()
```

```python
 data_tidy = data_melt.pivot(index=['Date', 'Drug_Name', 'time'],
                      columns='Parameter', values='reading').reset_index()
data_tidy.head()
```

```python
data_tidy.columns.name = None
```

```python
data_tidy.head()
```

```python
# compute a pivot day-wise average for each drug, for pressure and temperature

pd.pivot_table(data_tidy, index='Drug_Name', columns='Date', values=['Temperature'], aggfunc=np.mean )
```

```python
pd.pivot_table(data_tidy, index='Drug_Name', columns='Date', values=['Pressure'], aggfunc=np.mean )
```

```python
pd.pivot_table(data_tidy, index='Drug_Name', columns='Date', values=['Pressure', 'Temperature'], aggfunc=np.mean )
```

# ***Binning***

```python
temp_labels = ['low', 'medium', 'high', 'very_high']

temp_points = [5, 20, 35, 50, 60]

data_tidy['temp_cat'] = pd.cut(data_tidy['Temperature'], bins = temp_points, labels=temp_labels)

data_tidy.head()
```

```python
data_tidy['temp_cat'].value_counts()
```
