```python
import numpy as np

# Example array with NaN values
arr = np.array([3, np.nan, 2, 5, np.nan, 1])

# Sort array, NaNs are placed at the end
sorted_arr = np.sort(arr)
print(sorted_arr)  # Output: [ 1.  2.  3.  5. nan nan]

# Replace NaN with a large number before sorting (if you want NaN to be placed at the start)
arr_replaced = np.nan_to_num(arr, nan=float('-inf'))
sorted_arr_custom = np.sort(arr_replaced)
print(sorted_arr_custom)  # Output: [-inf   1.   2.   3.   5.   nan]
```

```python
import numpy as np

# Example array
arr = np.array([1, 3, 2, 5, 4])

# Sort by a custom comparison rule (e.g., sort by modulo 3 value)
indices = np.argsort(arr % 3)
sorted_arr_custom = arr[indices]
print(sorted_arr_custom)  # Output: [3 1 4 2 5]
```

```python
import numpy as np

# Example structured array
arr = np.array([(1, 'B'), (3, 'A'), (2, 'C')], dtype=[('num', int), ('char', 'U1')])

# Sort by 'num' field
sorted_arr = np.sort(arr, order='num')
print(sorted_arr)  # Output: [(1, 'B') (2, 'C') (3, 'A')]

# Sort by 'char' field
sorted_arr = np.sort(arr, order='char')
print(sorted_arr)  # Output: [(3, 'A') (1, 'B') (2, 'C')]
```

```python
# There is a condition for muliplying two matrices of shape (a, b) and (p, q)
# the condition is, that b == p

np.dot(a, c)
```

```python
# [(a, b) and (p, q) ] = (a, q) ; given  b == p

np.matmul(a,c)  # from R, Matlab, for making these people happy
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 2, 3], [4, 5, 6]])
# A.shape is (2, 2)
# B.shape is (2, 3)
C = np.dot(A, B)  # This is valid, resulting shape will be (2, 3)

D = np.array([[1, 2]])
# D.shape is (1, 2)
np.dot(A, D)  # This will raise a ValueError due to shape mismatch
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([1, 2])
# A.shape is (2, 2, 2)
# B.shape is (2,)
np.dot(A, B)
```

```python
A = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
B = np.array([1, 2])
# A.shape is (2, 2, 2)
# B.shape is (2,)
np.dot(A, B)  # This will raise a ValueError due to dimension mismatch
```
